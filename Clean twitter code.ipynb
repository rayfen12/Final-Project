## A function to tell if a tweet is a retweet or not

```python
def is_RT(tweet):
    if 'retweeted_status' not in tweet:
        return False      
    else:
        return True
```

## A function to tell what type of device is used to tweet

```python
#Create function for taking the most used Tweet sources off the #source column 
def device(tweet):
    if 'iPhone' in tweet['source'] or ('iOS' in tweet['source']):
        return 'iPhone'
    elif 'Android' in tweet['source']:
        return 'Android'
    elif 'Mobile' in tweet['source'] or ('App' in tweet['source']):
        return 'Mobile device'
    elif 'Mac' in tweet['source']:
        return 'Mac'
    elif 'Windows' in tweet['source']:
        return 'Windows'
    elif 'Bot' in tweet['source']:
        return 'Bot'
    elif 'Web' in tweet['source']:
        return 'Web'
    elif 'Instagram' in tweet['source']:
        return 'Instagram'
    elif 'Blackberry' in tweet['source']:
        return 'Blackberry'
    elif 'iPad' in tweet['source']:
        return 'iPad'
    else:
        return '-'
```

```python
# The code below is a for loop to input the metadata we need from the tweets into separate lists.
tw_list_date = []
tw_list_screen_name = []
tw_list_text = []
tw_list_retweet = []
tw_list_likes = []
tw_list_polarity = []
tw_list_location = []
tw_list_length = []
for element in tweet_data:
    tw_list_date.append(element['created_at'])
    tw_list_screen_name.append(element['user']['screen_name'])
    tw_list_text.append(element['full_text'])
    tw_list_retweet.append(element['retweet_count'])
    tw_list_likes.append(element['favorite_count'])
    analysis = TextBlob(element['full_text'])
    tw_list_polarity.append(analysis.sentiment.polarity)
    tw_list_location.append(element['user']['location'])
    tw_list_length.append(len(element['full_text']))
```

```python
tw_df = pd.DataFrame(list(zip(tw_list_date, tw_list_screen_name, tw_list_text, tw_list_length, tw_list_retweet, tw_list_likes, tw_list_location, tw_list_polarity)), 
               columns =['created_date', 'handle', 'text', 'length', 'retweet_count', 'likes_count', 'location', 'polarity']).astype({"created_date": 'datetime64'}).round({'polarity': 5})
```

```python
tw_df.head()
```

```python
# This for loop has the if conditions for polarity
sentiment_list = [] 
for value in tw_df["polarity"]: 
    if value == -1 and value <= -0.5: 
        sentiment_list.append("Negative") 
    elif value > -0.5 and value < 0: 
        sentiment_list.append("Somewhat Negative")
    elif value == 0:
        sentiment_list.append('Neutral')
    elif value >= 0.1 and value <= 0.5:
        sentiment_list.append('Somewhat Postive')
    else: 
        sentiment_list.append("Positive")
```

```python
tw_df['sentiment'] = sentiment_list
tw_df['device'] = list(map(device, tweet_data))
tw_df['RT'] = list(map(is_RT, tweet_data))
```

```python
#See the percentage of tweets that are #retweets:
RT_tweets = tw_df[tw_df['RT'] == True]
print(f"The percentage of retweets is {round(len(RT_tweets)/len(tw_df)*100)}% of all the tweets")
```

```python
#See how many tweets inside are plain text tweets (No RT or mention):
plain_text_tweets = tw_df[~tw_df['text'].str.contains("@")  & ~tw_df['text'].str.contains("RT") == False]
print(f"The percentage of plain text tweets is {round(len(plain_text_tweets)/len(tw_df)*100)}% of all the tweets")
```

```python
# The library below convert a collection of text documents to a matrix of token counts
from sklearn.feature_extraction.text import TfidfVectorizer

# The library stores dictionary values as counts 
from collections import Counter
```

```python
#Vectorizing the text column in the dataframe
vect = TfidfVectorizer(ngram_range=(2,5), stop_words='english')

summaries = "".join(tw_df['text'])
ngrams_summaries = vect.build_analyzer()(summaries)

word = Counter(ngrams_summaries).most_common(20)
```

## Creating the dataframe for the most common phrases in a user_timeline

```python
word_df = pd.DataFrame(word, columns=['common_words', 'word_count'])
```

```python
word_df.set_index('common_words', inplace=True)
```
